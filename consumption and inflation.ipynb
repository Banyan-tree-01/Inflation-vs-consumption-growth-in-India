```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("D:\PROJECTS\Consumption vs Inflation\Components of Gross Domestic Product.xlsx")
df
```

```python
df=df.dropna(axis=0)
```

```python
df
```

```python
import statsmodels.api as sm
```

```python
x=df['CPI All India General Index Combined (Base-2012=100)']
x=sm.add_constant(x)
y=df[' PFCE Constant Prices Base Year-2011-12']
model=sm.OLS(y,x).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as mplt
```

```python
mplt.scatter(df['CPI All India General Index Combined (Base-2012=100)'], df[' PFCE Constant Prices Base Year-2011-12'],color='blue',label="data points")
x=df['CPI All India General Index Combined (Base-2012=100)']
y=df[' PFCE Constant Prices Base Year-2011-12']
a=model.params['CPI All India General Index Combined (Base-2012=100)']
b=model.params['const']
mplt.plot(x,a*x+b,color='red',label='regression line')
mplt.xlabel("CPI Inflation (%)")
mplt.ylabel("Consumption Growth (%)")
mplt.title("Private Consumption Growth vs CPI Inflation")
mplt.grid(True)
mplt.show()
```

```python
#The regression results indicate that higher CPI inflation 
#is associated with a slightly lower real private consumption growth
#in India. However, the coefficient is not statistically significant at
#conventional levels (t = -0.382), suggesting weak evidence of a linear relationship
#in this specification. Given the small dataset (2013–14 to 2024–25), this result should be interpreted cautiously.
```
